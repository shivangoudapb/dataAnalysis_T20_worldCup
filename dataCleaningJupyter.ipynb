```python
import pandas as pd
import json
```

**match summary**

```python
with open('matchSummary.json') as f:
    data = json.load(f)
df_match=pd.DataFrame(data[0]['match_results'])
df_match.head()
```

```python
# Convert to datetime first
df_match["match_date"] = pd.to_datetime(df_match["match_date"])

# Extract only date
df_match["match_date"] = df_match["match_date"].dt.date

df_match.head()
```

```python
df_match.to_csv('match_summary.csv', index = False)
```

**batting summmary**

```python
with open('battingSummary.json') as f:
    data = json.load(f)
df_batting = pd.DataFrame(data[0]['battingSummary'])
df_batting.head()
```

```python
with open('battingSummary.json') as f:
    data = json.load(f)

    all_records = []

    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "not_out" if pd.isna(x) else "out")
df_batting.head(20)
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))

df_batting.head(15)
```

```python
match_ids_dict={}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]

match_ids_dict
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)

df_batting.head(15)
```

```python
df_batting.to_csv('batting_summary.csv', index = False)
```

**bowling summary**

```python
with open('bowlingSummary.json') as f:
    data = json.load(f)
df_bowling=pd.DataFrame(data[10]['bowlingSummary'])
df_bowling.head()
```

```python
with open('bowlingSummary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records
```

**player info**

```python
with open('playerInfo.json') as f:
    data = json.load(f)
data
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players.drop(columns=["input", "error"], inplace=True)

df_players.head(10)
```

```python
df_players["name"] = df_players["name"].astype(str)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('players_info.csv', index = False)
```
